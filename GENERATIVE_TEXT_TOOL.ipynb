```python
# Install required libraries (run only once)
!pip install transformers torch

# Import necessary libraries
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load GPT-Neo 125M model and tokenizer
model_name = "EleutherAI/gpt-neo-125M"

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Define text generation function
def generate_text(prompt, max_length=200, temperature=0.7, top_k=50, top_p=0.95):
    # Encode the prompt
    inputs = tokenizer(prompt, return_tensors="pt")

    # Generate output tokens
    outputs = model.generate(
        **inputs,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        do_sample=True,
        num_return_sequences=1
    )

    # Decode tokens into text
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text

# --- Example Usage ---
# Take user input
user_prompt = input("Enter a topic or prompt: ")

# Generate paragraph
generated_paragraph = generate_text(user_prompt)

# Print the generated text
print("\nGenerated Text:\n")
print(generated_paragraph)
```
